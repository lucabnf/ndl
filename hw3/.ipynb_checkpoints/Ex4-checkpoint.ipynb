```python
import numpy as np 
import matplotlib.pyplot as plt 
import networkx as nx 
from utility import *
from itertools import product
```

Using the presented algorithm, estimate the average degree $k$ and the disease spread parameters $β$ and $ρ$ for the pandemic.


```python
def PAM_graph(k, agents):
    
    # Starting point Graph 
    np.random.seed(0)
    G = nx.generators.classic.complete_graph(k + 1)

    # Preferential attachement
    for nt in range(k+1, agents):
        w = np.array(G.degree)[:,1]
        prob = w / np.sum(w) 
        c = (k + (nt % 2)) // 2

        idx = np.random.choice(G.nodes(), size =c, replace=False, p =prob)
        G.add_node(nt)

        for link in range(c):
            G.add_edge( nt, idx[link]) 
            
    return G
```

```python
def pandemic_simulation(G, beta, ro, initial_infected, pandemic_par ):
    
    # Pandemic simulation parameters
    vax =  pandemic_par["vax"]
    n_steps = pandemic_par["n_steps"]
    n_weeks = pandemic_par["n_weeks"]
    n_stati = pandemic_par["n_stati"]
    poli = pandemic_par["poli"]
    n_nodes = len(G)
    
    simu= np.zeros([n_steps,n_weeks, n_nodes])
    
    # Simulation loop
    for _ in range(0,n_steps):
        states = np.zeros([n_weeks, n_nodes])
                          
        # Initial condition 
        x0 = set_init_config(G, initial_infected, poli)
        vax_state = vaccination(vax[0], np.copy(x0), n_nodes)
        states[0] = vax_state
                          
        # Single simulation loop
        for i in range(1,n_weeks):    
            # Vaccination step
            perc = vax[i] - vax[i-1]
            vax_state = vaccination(perc, np.copy(states[i-1]), n_nodes)
                                    
            # Take last configuration
            x = np.copy( vax_state )
            rates = new_conf_rates(G, x, beta, ro, n_stati )
                          
            # Chose the next configuration 
            next_pos = np.argmax(np.cumsum(rates,axis=1) > np.random.rand(n_nodes,1), axis=1)
            states[i] = next_pos
                                   
        simu[_] = states
        
    return simu
                    
```

```python
def parameters_search(k0, beta0, ro0, delta_k, delta_beta, delta_ro, pandemic_par, I0, n_nodes):
    
    # Neighborhood generation
    k = [k0-delta_k, k0, k0+delta_k]
    beta = [beta0-delta_beta, beta0, beta0+delta_beta]
    ro = [ro0-delta_ro, ro0, ro0+delta_ro]
        
    cmb = list(product(k, beta, ro))    
    results = []
    
    # Neighborhood loop 
    for combination in cmb:
        
        # Graph Generation
        G = PAM_graph(combination[0], n_nodes )
        
        # Pandemic simulation
        simulation = pandemic_simulation(G, combination[1], combination[2], I0[0],  pandemic_par )
        
        # Compute the average number of newly infected individuals each week
        all_I = np.zeros([simulation.shape[0], simulation.shape[1]])

        for i in range( simulation.shape[0] ):     
            all_I[i] =  new_weekly( simulation[i], 1 ) 

        I = np.mean(all_I ,axis=0)
        
        # Compute RMSE 
        RMSE = np.sqrt( (np.sum(  (I[1:]-I0[1:])**2) ) / 15 )
        results.append(RMSE)
        
        print("[k:{}, beta:{}, ro:{}] --> RMSE:{}".format(combination[0], combination[1], combination[2],RMSE))

    # Chose the parameters that minimize the RMSE
    results = np.array(results)
    best_index = np.argmin(results)
    best_par = cmb[best_index]
    
    return best_par[0], best_par[1], best_par[2], results[best_index]
```

```python
# Pandemic parameters initialization 
pandemic_par = {
    "vax" : [5, 9, 16, 24, 32, 40, 47, 54, 59, 60, 60, 60, 60, 60, 60, 60],
    "n_steps" : 10,
    "n_weeks" : 15+1,
    "n_stati" : 4,
    "poli" : "random"
}

# Starting point for parameters to optimize
n_nodes = 934
I0 = [1, 1, 3, 5, 9, 17, 32, 32, 17, 5, 2, 1, 0, 0, 0, 0]
k, beta, ro = 10, 0.3, 0.6
delta_k, delta_beta, delta_ro = 1, 0.1, 0.1
```

```python
best_par =[k, beta, ro]
division_rmse = np.inf

# Loop of deltas reduction search
while True : 
    last_rmse = np.inf
    
    #best_local = best_par.copy()
    k, beta, ro = best_par[0],  best_par[1],  best_par[2]
    
    # Loop of around search
    while True  :
        k, beta, ro, new_rmse = parameters_search(k, beta, ro, delta_k, delta_beta, delta_ro, pandemic_par, I0, n_nodes)
    
        print("\n[k:{}, beta:{}, ro:{}] --> RMSE:{}\n".format(k, beta, ro, new_rmse))
    
        # If no bettere solution find go onc 
        if np.round(new_rmse, 2) >= np.round(last_rmse, 2):
            break
    
        last_rmse = new_rmse
        best_local = [k,beta,ro]
        
    
    # If no better solution found halve the deltasa
    if np.round(new_rmse, 2) >= np.round(division_rmse, 2):
            break
    
    delta_k = round(delta_k/2)
    delta_beta = delta_beta/2
    delta_ro = delta_ro/2
    
    print("New Deltas: [D_k: {}, D_beta:{}, D_ro:{}]".format(delta_k, delta_beta, delta_ro))

    division_rmse = last_rmse
    best_par = best_local
    
    print("New BEST {} --> RMSE: {}]".format(best_par, division_rmse))
```
